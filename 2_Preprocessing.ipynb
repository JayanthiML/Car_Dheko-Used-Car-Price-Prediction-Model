```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Raw_Data.csv')
```

```python
df.drop(['Unnamed: 0', 'it', 'ft', 'transmission', 'owner', 'priceActual', 'priceSaving', 'priceFixedText',
 'trendingText',  'Kms Driven', 'RTO', 'Ownership', 'Seats', 'Max Power', 'Torque', 'Displacement',
 'Seating Capacity', 'Engine'], axis=1, inplace=True)


df = df.rename(columns={'oem': 'Brand'})
df = df.rename(columns={'model': 'Model'})
df = df.rename(columns={'variantName': 'Variant Name'})
df = df.rename(columns={'modelYear': 'Model Year'})
df = df.rename(columns={'km': 'KMs Driven'})
df = df.rename(columns={'bt': 'Body Type'})
df = df.rename(columns={'ownerNo': 'Ownership'})
df = df.rename(columns={'Seats.1': 'Seating Capacity'})
df = df.rename(columns={'centralVariantId': 'Variant ID'})
df = df.rename(columns={'Max Power.1': 'Max Power'})
df = df.rename(columns={'price': 'Price'})
df = df.rename(columns={'No Door Numbers': 'Doors'})
```

```python
df['Body Type'] = df.apply(
    lambda row: 'MUV' if row['Variant Name'] == 'VIP Executive Lounge' and pd.isna(row['Body Type'])else row['Body Type'], axis=1)
df['Body Type'] = df.apply(
    lambda row: 'Minivans' if row['Variant Name'] == '5 Seater AC' and pd.isna(row['Body Type'])else row['Body Type'], axis=1)
df['Body Type'] = df.apply(
    lambda row: 'Minivans' if row['Variant Name'] == '5 Seater AC BSVI' and pd.isna(row['Body Type'])else row['Body Type'], axis=1)


df['Registration Year'] = df['Registration Year'].fillna(df['Model Year'])


df['Year of Manufacture'] = df['Year of Manufacture'].fillna(df['Model Year'])


df['Insurance Validity'] = df['Insurance Validity'].str.replace('Third Party insurance', 'Third Party')
df['Insurance Validity'] = df['Insurance Validity'].str.replace('1', 'First Party')
df['Insurance Validity'] = df['Insurance Validity'].str.replace('2', 'Second Party')
df['Insurance Validity'] = df['Insurance Validity'].fillna('Not Available')


df['Engine Displacement'] = df['Engine Displacement'].replace('0 cc', np.nan)
df['Engine Displacement'] = df.groupby('Variant Name')['Engine Displacement'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Engine Displacement'] = df.groupby('Model')['Engine Displacement'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Engine Displacement'] = df.groupby('Body Type')['Engine Displacement'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Mileage'] = df.groupby('Variant Name')['Mileage'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Mileage'] = df.groupby('Model')['Mileage'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Mileage'] = df.groupby('Engine Displacement')['Mileage'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df = df.dropna(subset=['Mileage'])


df['Engine Type'] = df.groupby('Variant Name')['Engine Type'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Engine Type'] = df.groupby('Model')['Engine Type'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Engine Type'] = df.groupby('Engine Displacement')['Engine Type'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Engine Type'] = df.groupby('Mileage')['Engine Type'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Seating Capacity'] = df.groupby('Body Type')['Seating Capacity'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Seating Capacity'] = df.groupby('Model')['Seating Capacity'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Seating Capacity'] = df.apply(
    lambda row: 7 if row['Variant Name'] == 'VIP Executive Lounge' and pd.isna(row['Seating Capacity'])else row['Seating Capacity'], axis=1)


df.drop(['Wheel Size', 'BoreX Stroke', 'Compression Ratio', 'Gross Weight', 'Front Tread', 'Rear Tread', 'Ground Clearance Unladen', 'Top Speed', 'Acceleration', 'Alloy Wheel Size'], axis=1, inplace=True)


df['Color'] = df['Color'].fillna(df['Color'].value_counts().index[2])


df['Max Power'] = df.groupby('Engine Displacement')['Max Power'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Max Power'] = df.groupby('Engine Type')['Max Power'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Max Torque'] = df.groupby('Engine Displacement')['Max Torque'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Max Torque'] = df.groupby('Engine Type')['Max Torque'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['No of Cylinder'] = df.groupby('Engine Type')['No of Cylinder'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['No of Cylinder'] = df.groupby('Engine Displacement')['No of Cylinder'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Values per Cylinder'] = df.groupby('Engine Type')['Values per Cylinder'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Values per Cylinder'] = df.groupby('Engine Displacement')['Values per Cylinder'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Value Configuration'] = df.groupby('Engine Type')['Value Configuration'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Value Configuration'] = df.groupby('Engine Displacement')['Value Configuration'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Value Configuration'] = df.groupby('Variant Name')['Value Configuration'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Value Configuration'] = df.groupby('Model')['Value Configuration'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Value Configuration'] = df.groupby('Brand')['Value Configuration'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


df['Fuel Suppy System'] = df.groupby('Engine Type')['Fuel Suppy System'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Fuel Suppy System'] = df.groupby('Engine Displacement')['Fuel Suppy System'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Fuel Suppy System'] = df.groupby('Variant Name')['Fuel Suppy System'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Fuel Suppy System'] = df.groupby('Model')['Fuel Suppy System'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['Fuel Suppy System'] = df.groupby('Brand')['Fuel Suppy System'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))


for i in df.columns[25:41]:
    df[i] = df.groupby('Variant Name')[i].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
    df[i] = df.groupby('Variant ID')[i].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
    df[i] = df.groupby('Model')[i].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
    df[i] = df.groupby('Brand')[i].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
    df[i] = df.groupby('Body Type')[i].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

```python
for column in df.columns:
    df[column] = df[column].astype(str)


df['Model'] = df.apply(lambda row: row['Model'].replace(row['Brand'], '').strip(), axis=1)


df['Model Year'] = pd.to_datetime(df['Model Year'], format='%Y', errors='coerce')
df['Model Year'] = df['Model Year'].dt.year

df['Year of Manufacture'] = df['Year of Manufacture'].astype(float).astype(int)
df['Year of Manufacture'] = pd.to_datetime(df['Year of Manufacture'], format='%Y', errors='coerce')
df['Year of Manufacture'] = df['Year of Manufacture'].dt.year

df['KMs Driven'] = df['KMs Driven'].str.replace(',', '').astype(int)


df['Ownership'] = df['Ownership'].astype(int)


df['Seating Capacity'] = df['Seating Capacity'].astype(float).astype(int)
df['No of Cylinder'] = df['No of Cylinder'].astype(float).astype(int)
df['Values per Cylinder'] = df['Values per Cylinder'].astype(float).astype(int)

df['City'] = df['City'].str.title()


df['Variant ID'] = df['Variant ID'].astype(int)


df['Registration Year'] = df['Registration Year'].str.extract(r'(\d+)')
df['Registration Year'] = pd.to_datetime(df['Registration Year'], format='%Y', errors='coerce')
df['Registration Year'] = df['Registration Year'].dt.year


df['Engine Displacement'] = df['Engine Displacement'].str.extract(r'(\d+)').astype(int)


df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.\d+|\d+)').astype(float)


df['Length'] = df['Length'].str.extract(r'(\d+)').astype(int)


df['Width'] = df['Width'].str.extract(r'(\d+)').astype(int)


df['Height'] = df['Height'].str.extract(r'(\d+)').astype(int)
df['Wheel Base'] = df['Wheel Base'].str.extract(r'(\d+)').astype(int)
df['Turning Radius'] = df['Turning Radius'].str.extract(r'(\d+)').astype(int)
df['Doors'] = df['Doors'].str.extract(r'(\d+)').astype(int)
df['Cargo Volumn'] = df['Cargo Volumn'].str.extract(r'(\d+)').astype(int)


import re

def clean_weight(weight):
    cleaned = re.sub(r'[^\d\-]', '', weight.replace(',', ''))

    if not cleaned:
        return None

    if '-' in cleaned:
        start, end = map(int, cleaned.split('-'))
        return int((start + end) / 2)
    else:
        return int(cleaned)

df['Kerb Weight'] = df['Kerb Weight'].apply(clean_weight)

df['Kerb Weight'] = df.groupby('Model')['Kerb Weight'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))

df['Kerb Weight'] = df['Kerb Weight'].astype(int)
```

```python
def price_convertion(price_str):
    price_str = price_str.replace('â‚¹', '').strip()
    parts = price_str.split()
    denomination = parts[-1].lower()  
    value = parts[0].replace(',', '')
    value = float(value)
    if denomination == 'lakh':
        value *= 100000
    elif denomination == 'crore':
        value *= 10000000
    else:
        value = value

    return int(value)


df['Price'] = df['Price'].apply(price_convertion)
```
